```python
mag = df.mag

count, bns = np.histogram(mag, bins=50)
rel_freq = count / mag.size
bin_centrds = (bns[1:] + bns[:-1]) / 2

plt.plot(bin_centrds, rel_freq);
plt.title('Earthquake magnitude vs. relative frequency in the US western coast.')
plt.ylabel('Relative frequency')
plt.xlabel(r'Magnitude ($\rm M_L$)');
```

As such, a more reasonable way to state the question would be as follows: A physical model that gives the predicted relative frequency of earthquakes of different magnitudes can be derived from physics and is given by a so-called *power-law*

where $P=10^x$. Show that taking the logarithm of both sides of this equation leads to a linear relationship between the earthquake magnitude $x$ and the log-frequency $\text{log}_{10}(y)$.

```python
log_counts = np.ma.log10(count).filled(0)

plt.plot(bin_centrds, log_counts);
plt.title('Earthquake magnitude vs. relative frequency in the US western coast.');
plt.ylabel('log10 event counts');
plt.xlabel(r'Magnitude ($\rm M_L$)');
```



Start with power function:<br>
$y = aP^b$<br>
where $P = 10 ^x$

Take log<sub>10</sub> of both sides of the equation:<br>
$\log(y) = \log(aP^b)$

Simplify:<br>
$\log(y) = \log(a) + \log(P^b)$

Simplify further:<br>
$\log(y) = \log(a) + b \log(P)$

Subsitute $10^x$ for $P$:<br>
$\log(y) = \log(a) + b \log(10^x)$

Resulting in:<br>
$\log(y) = \log(a) + bx)$


Now, our model takes the linear form $y = w_0 + w_1 x$<br>
where; $\log(y)$ is the prediction ($y$), $\log(a)$ is the intercept parameter ($w_0$), $b$ is the slope parameter ($w_1$), and $x$ is the feature ($x$).

Finally, **(40pts) use one of the methods that you developed for linear regression in class to fit a line to the log-transformed counts.  What is the slope of that line?  Is the model a good fit for all sizes of earthquake?**

!Answers go here

```python
from numpy.linalg import inv
d = 1
x = bin_centrds
y = log_counts

X = np.ones([len(x), 2])

X[:, 1] = x
XT = np.transpose(X)
XTX = np.dot(XT, X)
XTXi = np.linalg.inv(XTX)
XTy = np.dot(XT, y)

w = np.dot(XTXi, XTy)
m = w[1]
b = w[0]
x1 = np.linspace(x.min(), x.max())
ypred = m * x + b

plt.plot(bin_centrds, log_counts, label='Observed');
plt.plot(x1, ypred, label='Predicted')
plt.title('Earthquake magnitude vs. relative frequency in the US western coast.');
plt.ylabel('log10 event counts');
plt.xlabel(r'Magnitude ($\rm M_L$)');
plt.legend();
plt.ylim(0, 4)
plt.text(x=6, y=2, s='slope: %s' % round(m, 3));

print('The slope is %s' % round(m, 3))
```

```python
from scipy.stats import norm
resids = y - ypred
plt.hist(resids, bins=25);
plt.ylabel('count');
plt.xlabel('residual');
mean = np.mean(resids)
print('Residual Mean: %s' % mean)
```

The residuals appear to be reasonably normally distributed with a mean near zero, so assuming a linear relationship appears appropriate for modeling these data (based on a quick analysis). There is a left skew in the above histogram of residuals which coincides with the over estimation of number of earthquakes with magnitude > ~5.5 ML. However, the linear model appears to be well suited for predicting the number of earthquakes having magnitude < ~ 5.5 ML.
